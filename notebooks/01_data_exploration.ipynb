Phase 5: Churn Correlation Investigation

Objective: Determine if SmartAssist usage correlates with customer retention



```python
sma_users_df = smartassist_df.join(users[['user_id','account_id']].set_index('user_id'),on='user_id',how='left')
sma_users_df.head()
```

```python
accounts_df = sma_users_df.groupby('account_id').size().reset_index(name = "ai_event_count")
accounts_df
```

```python
account_ai_usage_df = subscriptions.merge(accounts_df, on="account_id", how = "left")
```

```python
account_ai_usage_df['ai_event_count'] = account_ai_usage_df['ai_event_count'].fillna(0)
account_ai_usage_df.head()
```

```python
def classify_accounts(count):
    if count == 0:
        return "No AI usage"
    elif count>0 and count<10:
        return "Light AI usage"
    else:
        return "Heavy AI usage"

account_ai_usage_df['AI_activity_type'] = account_ai_usage_df['ai_event_count'].apply(classify_accounts)
```

```python
account_ai_usage_df[['account_id','AI_activity_type']]
```

```python
account_ai_usage_df
```

```python
account_ai_usage_summary = account_ai_usage_df.groupby('AI_activity_type').agg(Total_accounts =('account_id','count') 
                                                    , churned_accounts =('status',lambda x: (x.str.lower() == 'churned').sum())
                                                    , Average_mrr_per_account = ('mrr','mean')
                                                    ).reset_index()
account_ai_usage_summary
```

```python
account_ai_usage_summary['churn_rate'] = round((account_ai_usage_summary['churned_accounts']/account_ai_usage_summary['Total_accounts'])*100,2)
account_ai_usage_summary
```

```python
overall_churn_rate = (subscriptions['status'].str.lower()=='churned').sum()/subscriptions['account_id'].nunique()*100
```

```python
print(f"Overall Churn Rate: {overall_churn_rate:.2f}%")
```

```python
account_ai_usage_summary['churn_rate_ratio'] = account_ai_usage_summary['churn_rate'] / overall_churn_rate
account_ai_usage_summary.sort_values('churn_rate_ratio',ascending=False)
```

```python
plt.Figure(figsize=(8,5))
ax = account_ai_usage_summary.plot(kind='bar',x='AI_activity_type',y="churn_rate",xlabel = "AI Usage Segment", ylabel='Churn rate (%)')

plt.axhline(
    overall_churn_rate,
    linestyle="--",
    label=f"Overall churn rate ({overall_churn_rate:.2f}%)"
)
for i, row in account_ai_usage_summary.iterrows():
    plt.text(
        i,
        row["churn_rate"] + 0.3,
        f"n={row['Total_accounts']}",
        ha="center"
    )
plt.legend()
plt.show()
```

Phase 6: Support Cost Analysis

Objective: Quantify the hidden costs of SmartAssist

```python
support_tickets.head()
```

```python
ai_keywords = r"smartassist|smart assist\bai\b|ai"
support_tickets['is_ai_related']=(support_tickets['ticket_description'].str.contains(ai_keywords)) | (support_tickets['category'].str.lower() == 'smartassist')
```

```python
support_tickets.head()
```

```python
support_tickets['created_date']=pd.to_datetime(support_tickets['created_date'])
support_tickets
```

```python
support_tickets["month"] = support_tickets['created_date'].dt.to_period("M").astype(str)
```

```python
monthly_tickets_trend = support_tickets.groupby('month').agg(total_tickets = ('ticket_id','count'), ai_tickets = ('is_ai_related','sum')).reset_index()
```

```python
monthly_tickets_trend
```

```python
monthly_tickets_trend['ai_percentage'] =  round((monthly_tickets_trend['ai_tickets']*100 / monthly_tickets_trend['total_tickets']),2)
```

```python
monthly_tickets_trend
```

```python
resolution_comparison = support_tickets.groupby('is_ai_related')['resolution_time_hours'].mean().reset_index()

resolution_comparison['ticket_type'] = resolution_comparison['is_ai_related'].map({False:'Not AI related' , True:'AI related'})
resolution_comparison
```

```python
ai_resolution = resolution_comparison.loc[
    resolution_comparison["is_ai_related"] == True,
    "resolution_time_hours"
].values[0]

non_ai_resolution = resolution_comparison.loc[
    resolution_comparison["is_ai_related"] == False,
    "resolution_time_hours"
].values[0]

resolution_ratio = ai_resolution / non_ai_resolution
```

```python
resolution_ratio
print(f"AI tickets take {resolution_ratio:.2f}x longer to resolve than non-AI tickets.")
```

```python
csat_comparison = support_tickets.groupby('is_ai_related').agg(avg_csat_score=("csat_score", "mean")).reset_index()

csat_comparison['ticket_type'] = csat_comparison['is_ai_related'].map({False:'Not AI related' , True:'AI related'})
csat_comparison
csat_comparison
```

```python
escalation_comparison = support_tickets.groupby("is_ai_related").agg(
    escalation_rate=("escalated", "mean")
).reset_index()

escalation_comparison["escalation_rate"] = escalation_comparison["escalation_rate"] * 100

escalation_comparison["ticket_type"] = escalation_comparison["is_ai_related"].map({
    True: "AI-Related",
    False: "Non-AI"
})

escalation_comparison
```

```python
plt.Figure(figsize=(8,5))
plt.plot(monthly_tickets_trend["month"], monthly_tickets_trend["total_tickets"], marker="o", label="Total Tickets")
plt.plot(monthly_tickets_trend["month"], monthly_tickets_trend["ai_tickets"], marker="o", label="AI-Related Tickets")

plt.title("AI-Related Support Ticket Trend")
plt.xlabel("Month")
plt.ylabel("Number of Tickets")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

The number of AI-related tickets grew from roughly 360 in August to around 750 by October and remained high through December. Cost Implications:
The rapid growth in AI-related tickets suggests that SmartAssist is creating a meaningful support burden for the team.
