```python
new = users.join(subscriptions.set_index('account_id'), on='account_id', how='left')
new.head()
```

```python
new['smartassist_users'] = new['user_id'].isin(smartassist_df['user_id'])
new.head()
```

```python
tier_summary = new.groupby('plan_tier').agg(total_users=('user_id','count'), sma_users = ('smartassist_users','sum')).reset_index()
```
