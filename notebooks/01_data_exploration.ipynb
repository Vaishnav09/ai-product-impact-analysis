```python
def classify_range(count):
    if count == 1:
        return "One-Time User"
    elif count>1 and count<10:
        return "Light Repeat Users"
    else:
        return "Heavy Users"


uses['User Type'] = uses['count'].apply(classify_range)
```

```python
Users_with_2plus_usage =  uses[uses['User Type'].isin(['Heavy Users','Light Repeat Users'])]['user_id'].count()
```

Time Based Analysis

For repeat users, calculate days between first and last SmartAssist usage

How long do users stay engaged with the feature?


```python
sma_uses_df = smartassist_df.join(uses.set_index('user_id'),on='user_id',how='left')
```

```python
sma_uses_df.head()
```

```python
repeat_uses_df = sma_uses_df[sma_uses_df['User Type'].isin(['Heavy Users','Light Repeat Users'])]
repeat_uses_df.info()
```

```python
repeat_uses_df['event_date'] = pd.to_datetime(repeat_uses_df['event_date'])
repeat_uses_df.info()
```

```python
users_engage_df = repeat_uses_df.groupby('user_id').agg(first_use = ('event_date','min'),last_use = ('event_date','max')).reset_index()
```

```python
users_engage_df['duration(in days)'] = (users_engage_df['last_use'] - users_engage_df['first_use']).dt.days

users_engage_df
```

```python
usage_summary = sma_uses_df.groupby('User Type')['user_id'].nunique().reset_index()

usage_summary = usage_summary.rename(columns = {'user_id':'user_counts'})

```

```python
usage_summary['percent'] = round((usage_summary['user_counts']/sma_uses_df['user_id'].nunique())*100,2)

usage_summary
```

```python
plt.Figure(figsize=(6,4))
colors = ["red","orange","green"]

ax = usage_summary.sort_values('user_counts').plot(kind='barh',x='User Type',y='user_counts',ylabel='User Segments', xlabel='#Users',color = colors)
for i,val in enumerate(usage_summary['user_counts']):
    p = usage_summary['percent'].iloc[i]
    plt.text(val+10,i,f"{p}%", va="center")
plt.title("SmartAssist Usage Distribution")
plt.show()
```

Phase 5: Churn Correlation Investigation

Objective: Determine if SmartAssist usage correlates with customer retention



```python
sma_users_df = smartassist_df.join(users[['user_id','account_id']].set_index('user_id'),on='user_id',how='left')
sma_users_df.head()
```

```python
accounts_df = sma_users_df.groupby('account_id').size().reset_index(name = "ai_event_count")
accounts_df
```

```python
account_ai_usage_df = subscriptions.merge(accounts_df, on="account_id", how = "left")
```

```python
account_ai_usage_df['ai_event_count'] = account_ai_usage_df['ai_event_count'].fillna(0)
account_ai_usage_df.head()
```

```python
def classify_accounts(count):
    if count == 0:
        return "No AI usage"
    elif count>0 and count<10:
        return "Light AI usage"
    else:
        return "Heavy AI usage"

account_ai_usage_df['AI_activity_type'] = account_ai_usage_df['ai_event_count'].apply(classify_accounts)
```

```python
account_ai_usage_df[['account_id','AI_activity_type']]
```

```python
account_ai_usage_df
```

```python
account_ai_usage_summary = account_ai_usage_df.groupby('AI_activity_type').agg(Total_accounts =('account_id','count') 
                                                    , churned_accounts =('status',lambda x: (x.str.lower() == 'churned').sum())
                                                    , Average_mrr_per_account = ('mrr','mean')
                                                    ).reset_index()
account_ai_usage_summary
```

```python
account_ai_usage_summary['churn_rate'] = round((account_ai_usage_summary['churned_accounts']/account_ai_usage_summary['Total_accounts'])*100,2)
account_ai_usage_summary
```

```python
overall_churn_rate = (subscriptions['status'].str.lower()=='churned').sum()/subscriptions['account_id'].nunique()*100
```

```python
print(f"Overall Churn Rate: {overall_churn_rate:.2f}%")
```

```python
account_ai_usage_summary['churn_rate_ratio'] = account_ai_usage_summary['churn_rate'] / overall_churn_rate
account_ai_usage_summary.sort_values('churn_rate_ratio',ascending=False)
```

```python
plt.Figure(figsize=(8,5))
ax = account_ai_usage_summary.plot(kind='bar',x='AI_activity_type',y="churn_rate",xlabel = "AI Usage Segment", ylabel='Churn rate (%)')

plt.axhline(
    overall_churn_rate,
    linestyle="--",
    label=f"Overall churn rate ({overall_churn_rate:.2f}%)"
)
for i, row in account_ai_usage_summary.iterrows():
    plt.text(
        i,
        row["churn_rate"] + 0.3,
        f"n={row['Total_accounts']}",
        ha="center"
    )
plt.legend()
plt.show()
```
